```python
import numpy as np

from scipy.stats import gaussian_kde

import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.collections import PolyCollection
```

```python
pred_arr = np.load("/media/zhongying/Elements/Zhongying/PM2.5_estimation_US/model/saved_model/pm25/bilstm/Western_model_1_pred.npy")
truth_arr = np.load("/media/zhongying/Elements/Zhongying/PM2.5_estimation_US/model/saved_model/pm25/bilstm/Western_model_1_truth.npy")

pred_arr = pred_arr.squeeze()
truth_arr = truth_arr.squeeze()
```

```python
plt.figure(figsize=(10,10))
xy = np.vstack([pred_arr, truth_arr])
z = gaussian_kde(xy)(xy)
plt.scatter(pred_arr, truth_arr, c=z)
# plt.yscale('log')
# plt.xscale('log')

p1 = max(max(pred_arr), max(truth_arr))
p2 = min(min(pred_arr), min(truth_arr))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('True Values', fontsize=15)

plt.ylabel('Predictions', fontsize=15)
plt.axis('equal')
plt.title('PM2.5 Predicted VS True Values \n BiLSTM', fontsize=20)
plt.show()
```

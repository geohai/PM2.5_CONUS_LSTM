```python
import xarray
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Visualize Merged Xarray 

```python
date = '2021-07-01'

merged_xr = xarray.open_dataset(f'../data/merged_inputs/{date}_with_imputation.nc')

merged_xr
```

```python
f, ax1 = plt.subplots(1, 1, figsize=(12, 6))

# The first plot (in kelvins) chooses "viridis" and uses the data's min/max
merged_xr['aod_047'].plot(ax=ax1, 
                          cbar_kwargs={"label": "Var"},
#                           vmin=0, vmax=35
                         )
ax1.set_title(f"Var - {date}")
```

```python
f, ax1 = plt.subplots(1, 1, figsize=(12, 6))

# The first plot (in kelvins) chooses "viridis" and uses the data's min/max
(merged_xr['knnidw_pm25'].notnull()).plot(ax=ax1, 
                          cbar_kwargs={"label": "Var"},
#                           vmin=0, vmax=35
                         )
ax1.set_title(f"Var - {date}")
```
